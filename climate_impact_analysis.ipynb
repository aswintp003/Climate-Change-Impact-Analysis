```python
df[df[['AvgTemp', 'AvgTempUncertainty']].isnull().any(axis=1)]    # selects rows from df where there is missing value
```

```python
df['AvgTemp_missing'] = df['AvgTemp'].isnull().astype(int)  
```

```python
df['AvgTemp'] = df.groupby('Season')['AvgTemp'].transform(lambda x: x.fillna(x.mean()))
```

```python
df[df['AvgTemp_missing'] == 1]
```

```python
df['AvgTempUncertainty_missing'] = df['AvgTempUncertainty'].isnull().astype(int)
```

```python
df['AvgTempUncertainty'] = df.groupby('Season')['AvgTempUncertainty'].transform(lambda x: x.fillna(x.mean()))
```

```python
df[df['AvgTempUncertainty_missing'] == 1]
```

```python
df.drop(columns=['AvgTemp_missing', 'AvgTempUncertainty_missing'], inplace=True)
```

```python
df.head()
```

```python
df['CDate'] = pd.to_datetime(df['CDate'], errors='coerce')  # to convert column named 'CDate' into datetime objects
```

```python
df['CDate'].dtype # to check the datatype of cdate
```

```python
# to convert a string representation of a geographical coordinate (latitude or longitude) into a numerical float value, 
# handling cardinal directions (N, S, E, W) and missing values.

def convert_geo(val):
    if pd.isna(val): return None
    val = val.strip().upper()
    if val[-1] in ['N', 'E']: return float(val[:-1])
    elif val[-1] in ['S', 'W']: return -float(val[:-1])
    return None
```

```python
df['Latitude'] = df['Latitude'].apply(convert_geo)
```

```python
df['Longitude'] = df['Longitude'].apply(convert_geo)
```

```python
df['Latitude'].dtype
```

```python
# This shows how temperatures vary across seasons.

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Season', y='AvgTemp')
plt.title("Temperature Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Average Temperature")
plt.show()
```

```python
# This helps you see if rainfall levels affect crop success

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Label', y='Precipitation')
plt.title("Precipitation vs Crop Outcome")
plt.xlabel("Crop Label (0 = Fail, 1 = Success)")
plt.ylabel("Precipitation")
plt.show()
```

```python
# This plot shows how average temperature changes month-over-month, with separate lines for each season to highlight seasonal trends.

df['Month'] = df['CDate'].dt.month

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Month', y='AvgTemp', hue='Season', marker='o')
plt.title('Average Temperature Over Time by Season')
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.show()
```

```python
# Impact of weather on crop classification (box plot)
# This plot compares the distribution of average temperature for each Label group, which could be used to differentiate between successful (0) and unsuccessful (1) crop outcomes.

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Label', y='AvgTemp')
plt.title('Average Temperature Distribution by Crop Label')
plt.xlabel('Crop Label (0 or 1)')
plt.ylabel('Average Temperature')
plt.show()
```

```python
# Sort crops by median temperature

ordered_crops = df.groupby('Crop')['AvgTemp'].median().sort_values().index

plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Crop', y='AvgTemp', order=ordered_crops)
plt.xticks(rotation=45)
plt.title("Temperature Distribution by Crop Type (Sorted by Median)")
plt.xlabel("Crop")
plt.ylabel("Average Temperature")
plt.show()
```
